# Data Visualization

```python
#To determine the highest number of confirmed cases in WHO region ,we shall plot it in bar diagram.
df_grouped=df.groupby('WHO Region')['Confirmed'].sum().reset_index()

plt.figure(figsize=(10,6))
plt.bar(df_grouped['WHO Region'],df_grouped['Confirmed'],color='green')
plt.xlabel('WHO regions')
plt.ylabel('Confirmed Cases')
plt.title('Confirmed Cases per WHO region')
plt.tight_layout()
plt.show()

#To determine the highest number of deaths in WHO region
df_death=df.groupby('WHO Region')['Deaths'].sum().reset_index()

plt.figure(figsize=(10,6))
plt.bar(df_death['WHO Region'],df_death['Deaths'],color='skyblue')
plt.xlabel('WHO regions')
plt.ylabel('Death Cases')
plt.title('Death Cases per WHO region')
plt.tight_layout()
plt.show()

#To determine the highest number of recovered cases in WHO region
df_death=df.groupby('WHO Region')['Recovered'].sum().reset_index()

plt.figure(figsize=(10,6))
plt.bar(df_death['WHO Region'],df_death['Recovered'],color='red')
plt.xlabel('WHO regions')
plt.ylabel('Recovered Cases')
plt.title('Recovered Cases per WHO region')
plt.tight_layout()
plt.show()
```

####  America has the largest number of confirmed cases and Western Pacific has the least. 
####                                      America has the largest number of deaths as well as recovered cases.

```python
# To determine which country has the maximum number of confirmed cases.
max_cases=df.loc[df['Confirmed'].idxmax()]
print('Country with the maximum confirmed cases')
print(max_cases['Country/Region'])
print('Number of confirmed cases:',max_cases['Confirmed'])
```

```python
# Sorting the top 5 countries with the most number of confirmed cases
df=df.sort_values(by='Confirmed',ascending=False)
top_5=df.head(5)
top_5
```

```python
heatmap_data = top_5.set_index('Country/Region')['Confirmed'].reset_index()

# Reshape the data for seaborn heatmap
heatmap_data = heatmap_data.pivot_table(index=None, columns='Country/Region', values='Confirmed')

# Plotting the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, annot=True, fmt=".0f", cmap='YlOrRd', linewidths=0.5)
plt.title('Top 5 Countries with Most Confirmed COVID-19 Cases')
plt.xlabel('Country')
plt.ylabel('Confirmed Cases')
plt.show()
```

```python
# Visualizing the percentage increase in a week of Covid cases among the WHO regions in a bar graph.

# Grouping the data by WHO region and calculating the mean for each group
grouped = df.groupby('WHO Region')['1 week % increase'].mean().reset_index()

# Plotting the bar chart
plt.figure(figsize=(10, 6))
plt.bar(grouped['WHO Region'], grouped['1 week % increase'], color='blue')
plt.xlabel('WHO Region')
plt.ylabel('1 Week Percentage Increase')
plt.title('1 Week Percentage Increase in COVID-19 Cases by WHO Region')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

#### Western Pacific Region has the most increasing Covid-cases during the interval of a week.
